```python
shares = []
figura = plt.figure(figsize= (16,7))
for dia in [' weekday_is_monday', ' weekday_is_tuesday', ' weekday_is_wednesday',' weekday_is_thursday', ' weekday_is_friday', ' weekday_is_saturday',
       ' weekday_is_sunday']:
    shares.append(data[data[dia]==1][' shares'].median())
ax = sns.barplot(x= ['segunda','terÃ§a','quarta','quinta','sexta','sabado','domingo'], y= shares)
ax.set(xlabel='Dias da semana', ylabel='media de compartilhamento')
plt.show()
```

```python
categorias  = []
figura = plt.figure(figsize= (16,7))
for tema in [' data_channel_is_lifestyle',
       ' data_channel_is_entertainment', ' data_channel_is_bus',
       ' data_channel_is_socmed', ' data_channel_is_tech',
       ' data_channel_is_world']:
    categorias.append(data[data[tema]==1][' shares'].median())

    
ax = sns.barplot(x= ['Estilo de vida',
       'entretenimento', 'transporte',
       'midias sociais ', 'tecnologia',
       'mundo'], y= categorias)
ax.set(xlabel='categorias', ylabel='media de compartilhamento')

plt.show()
```
